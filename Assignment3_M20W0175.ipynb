```python
# Split to train and test 7:3
train_size = int(len(df) * 0.7)
test_size = len(df) - train_size
train, test = df.iloc[:train_size], df.iloc[train_size:]
print(train.shape, test.shape)
```

```python
# Plot data
plt.plot(df, label="birth rate")
plt.axvline(x=test.iloc[0].name, color="orange")  # mark starting point of the test subset
plt.legend()
plt.show()
```

```python
# Normalize data
scaler = StandardScaler()
scaler = scaler.fit(train[["birth_rate"]])

trainT = scaler.transform(train[["birth_rate"]])
testT = scaler.transform(test[["birth_rate"]])

# np.array(train["hr"]).reshape(-1, 1).shape
# equals to
# train[["hr"]].shape
```

```python
def create_dataset(x, y, time_steps=1):
    xs, ys = [], []
    for i in range(len(x) - time_steps):
        xs.append(x[i:(i + time_steps)])
        ys.append(y[i:(i + time_steps)])
    return np.array(xs), np.array(ys)
```

```python
TIME_STEPS = 1

# reshape to [samples, time_steps, n_features]
x_train, y_train = create_dataset(trainT, trainT, TIME_STEPS)
x_test, y_test = create_dataset(testT, testT, TIME_STEPS)
print(x_train.shape)
print(y_train.shape)
```

```python
model = keras.Sequential()
model.add(keras.layers.LSTM(
    units=4,
    input_shape=(x_train.shape[1], x_train.shape[2])
))
# model.add(keras.layers.Dropout(rate=0.2))
model.add(keras.layers.RepeatVector(n=x_train.shape[1]))
model.add(keras.layers.LSTM(units=4, return_sequences=True))
# model.add(keras.layers.Dropout(rate=0.2))
model.add(keras.layers.TimeDistributed(keras.layers.Dense(units=x_train.shape[2])))
model.compile(loss="mae", optimizer="adam")
model.summary()
```

```python
history = model.fit(
    x=x_train,
    y=y_train,
    epochs=100,
    batch_size=16,
    validation_split=0.1,
    shuffle=False
)
```

```python
# PLOT LOSS
plt.plot(history.history["loss"], label="train loss")
plt.plot(history.history["val_loss"], label="validation loss")
plt.legend()
plt.show()
```

```python
# CHECK LOSS ON TRAIN SET
X_train_pred = model.predict(x_train)
train_mae_loss = np.mean(np.abs(X_train_pred - x_train), axis=1)
sns.distplot(train_mae_loss, bins=30, kde=True)
plt.show()
```

```python
# CHECK LOSS ON TEST SET
X_test_pred = model.predict(x_test)
test_mae_loss = np.mean(np.abs(X_test_pred - x_test), axis=1)
```

```python
THRESHOLD = 0.6

test_score_df = pd.DataFrame(index=test[:-TIME_STEPS].index)
test_score_df["loss"] = test_mae_loss
test_score_df["threshold"] = THRESHOLD
test_score_df["anomaly"] = test_score_df.loss > test_score_df.threshold
test_score_df["birth_rate"] = testT[:-TIME_STEPS]
test_score_df.head()

anomalies = test_score_df[test_score_df.anomaly == True]
not_anomalies = test_score_df[test_score_df.anomaly == False]
anomalies.head()
```

```python
# PLOT LOSS
plt.plot(test_score_df.index, test_score_df.loss, label="loss")
plt.plot(test_score_df.index, test_score_df.threshold, label="threshold")
sns.scatterplot(
    x=anomalies.index,
    y=anomalies.loss,
    color=sns.color_palette()[3],
    s=150,
    label="an anomaly"
)
sns.scatterplot(
    x=not_anomalies.index,
    y=not_anomalies.loss,
    color=sns.color_palette()[2],
    s=150,
    label="not an anomaly"
)
# plt.xticks(rotation=25)
plt.legend()
plt.show()
```

```python
# PLOT ANOMALIES
fig, ax = plt.subplots(1)
plt.plot(
    test[:-TIME_STEPS].index,
    scaler.inverse_transform(testT[:-TIME_STEPS]),
    label="birth rate"
)
ax.set_ylabel("birth rate")
sns.scatterplot(
    x=anomalies.index,
    y=scaler.inverse_transform(anomalies[["birth_rate"]]).reshape(-1),
    color=sns.color_palette()[3],
    s=150,
    label="an anomaly"
)
sns.scatterplot(
    x=not_anomalies.index,
    y=scaler.inverse_transform(not_anomalies[["birth_rate"]]).reshape(-1),
    color=sns.color_palette()[2],
    s=150,
    label="not an anomaly"
)
# plt.xticks(rotation=25)
plt.legend()
plt.show()
```
